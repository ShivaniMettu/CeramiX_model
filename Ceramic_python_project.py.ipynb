```python
# splitting 
from sklearn.model_selection import train_test_split
x=data.iloc[:,2:20]
x
```

```python
y=data.iloc[:,1]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
# model 1: Logestic regression
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=lr.predict(x_test)
lrAccuracy=accuracy_score(y_pred,y_test)
print("Accuracy: ",lrAccuracy)
```

```python
# Function to predict based on dynamic input
def predict_dynamic_input(model, input_data):
    # Convert input data to DataFrame
    input_df = pd.DataFrame([input_data], columns=x.columns)
    
    # Make prediction
    prediction = model.predict(input_df)
    
    return prediction[0]

# Example dynamic input (replace with actual input values)
dynamic_input = {
    'Na2O': 0.60, 'MgO': 0.39, 'Al2O3': 19.65, 'SiO2': 71.95, 'K2O': 4.82, 'CaO': 0.33,
    'TiO2': 0.17, 'Fe2O3': 1.11, 'MnO': 633, 'CuO': 12, 'ZnO': 74, 'PbO2': 13,
    'Rb2O': 432, 'SrO': 0.5, 'Y2O3': 43, 'ZrO2': 82, 'P2O5': 92
}

# Predict using dynamic input
prediction = predict_dynamic_input(lr, dynamic_input)
print("Prediction for dynamic input: ", prediction)
```
